### Here the target column is diagnosis where it labels the particular cancer info as malignant or benign

```python
from sklearn.model_selection import train_test_split
# Splitting the data into training and testing data
X_train, X_test, y_train, y_test = train_test_split(df.iloc[:,1:],df.iloc[:,0],test_size=0.2,random_state=42)
```

```python
X_train
```

```python
#converting all features to a common scale
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
X_train = scaler.fit_transform(X_train)
X_test = scaler.fit_transform(X_test)
```

```python
X_train
```

```python
#USing KNN algorithm
from sklearn.neighbors import KNeighborsClassifier

#k = 3 at first and after cross validation changed it to 9
knn = KNeighborsClassifier(n_neighbors = 9)
```

```python
knn.fit(X_train,y_train)
```

```python
y_pred = knn.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

### Applying cross validation to find the best 'k' value

```python
scores = []

for i in range(1,16):
    knn = KNeighborsClassifier(n_neighbors = i)
    knn.fit(X_train, y_train)
    y_pred = knn.predict(X_test)
    scores.append(accuracy_score(y_test,y_pred))
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(range(1,16),scores)
```

#### Here the highest accuracy is greater than .96 which is for 9. So Lets use it as k for our previous model building.
