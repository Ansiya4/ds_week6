```python
# Import libraries
import numpy as np
import pandas as pd

# Import one-hot encoder
from sklearn.preprocessing import OneHotEncoder
```

```python
# Import dataset
df = pd.read_csv('data/Iris.csv')

# Create one-hot encoder instance
onehot_encoder = OneHotEncoder()
```

```python
df
```

```python
df['Species'].unique()
```

```python
onehot= onehot_encoder.fit_transform(df[['Species']]).toarray()
```

```python
onehot
```

```python
onehot_columns = onehot_encoder.get_feature_names_out()
```

```python
onehot_columns
```

```python
df[onehot_columns] = onehot
```
